```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("housing.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# ========================
# 2. Top 5 Most Frequent Values in Selected Columns
# ========================
selected_columns = ["housing_median_age", "median_house_value"]

for col in selected_columns:
    print(f"\nTop 5 most frequent values in {col}:")
    print(df[col].value_counts().head(5))
```

```python
# 3. Distribution Plots
plt.figure(figsize=(8, 4))
sns.histplot(df["median_house_value"], kde=True, bins=30)
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot for duration
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["housing_median_age"])
plt.title("Boxplot of Housing Median Age")
plt.show()
```
